```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
from sklearn.model_selection import cross_val_predict
import numpy as np
```

```python
#Creer des données synthetiques 

N = 1000
p0 = 0.05
theta0 = 0
wt = np.pi/32
theta = np.random.randn(N)*np.pi/24 
a = (theta-theta0)/wt
p = p0/2 + ((1-p0)/(1+np.exp(-a)))

#reshape 

p = np.array(p).reshape(-1, 1)
theta = np.array(theta).reshape(-1,1)

#regression logistique 

model = linear_model.LinearRegression()

#model = LogisticRegression (C = 3, max_iter = 100, penalty = 'l2')  

# p prediction error 
predicted = cross_val_predict(model, theta, p, cv = 2)

#plot

fig, ax = plt.subplots()
ax.scatter(p, predicted, edgecolors=(0, 0, 0))
plt.plot([p.min(), p.max()], [p.min(),p.max()], 'k--', lw=4)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.title("p prediction error")
plt.show()
```

```python
# theta prediction error

predicted = cross_val_predict(model, p, theta, cv = 2)

fig, ax = plt.subplots()
ax.scatter(theta, predicted, edgecolors=(0, 0, 0))
plt.plot([theta.min(), theta.max()], [theta.min(),theta.max()], 'k--', lw=4)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.title("theta prediction error")
plt.show()
```
